```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming we have the file named "cpu_frequencies.csv"
file_path = './cpu_frequencies.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Drop the 'Time' column as it's not needed for the computation
df = df.drop(columns=['Time'])
df = df.iloc[300:400]

# Plot each column's data
plt.figure(figsize=(20, 8))

for column in df.columns:
    plt.plot(df.index, df[column], label=f'Core {column}')

plt.xlabel('Time Index')
plt.ylabel('Core Frequency (GHz)')
plt.title('CPU Core Frequencies Over Time')
plt.legend(loc='best', ncol=2)
plt.grid(True)
plt.show()

```

```python
# Assuming the file is available with the specified path
csv_file_path = '../data/ecp_power_res/no_power_shift/uncore_freq/bert_large_uncore_freq.csv'

# Load the CSV file into a DataFrame
df_uncore = pd.read_csv(csv_file_path)

# Plotting the uncore frequency
plt.figure(figsize=(10, 6))
plt.plot(df_uncore.index, df_uncore['Core 0 Uncore Frequency (GHz)'],color="red")
# plt.title('Uncore Frequency Over Time')
plt.xlabel('Time Index')
plt.ylabel('Uncore Frequency (GHz)')
plt.grid(True)
plt.show()
```
