```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming we have the file named "cpu_frequencies.csv"
file_path = './cpu_frequencies.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Drop the 'Time' column as it's not needed for the computation
df = df.drop(columns=['Time'])

# Function to calculate the average of values greater than 0.8 in each row
def average_greater_than_08(row):
    filtered_values = row[row > 0.8]
    if len(filtered_values) > 0:
        return filtered_values.mean()
    else:
        return 0.8  # If no values > 0.8, return 0

# Apply the function to each row
df['Average > 0.8'] = df.apply(average_greater_than_08, axis=1)

# Plot the average values
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Average > 0.8'])
plt.title('Average CPU Core Frequency')
plt.xlabel('Time index')
plt.ylabel('Average Frequency (GHz)')
plt.grid(True)
plt.show()
```

```python
# Assuming the file is available with the specified path
csv_file_path = '../data/ecp_power_res/no_power_shift/uncore_freq/bert_large_uncore_freq.csv'

# Load the CSV file into a DataFrame
df_uncore = pd.read_csv(csv_file_path)

# Plotting the uncore frequency
plt.figure(figsize=(10, 6))
plt.plot(df_uncore.index, df_uncore['Core 0 Uncore Frequency (GHz)'],color="red")
plt.title('Uncore Frequency Over Time')
plt.xlabel('Time Index')
plt.ylabel('Uncore Frequency (GHz)')
plt.grid(True)
plt.show()
```
