```python
import csv
import matplotlib.pyplot as plt

def plot_results(csv_file):
    gpu_clocks = []
    max_powers = []

    # Read data from CSV file
    with open(csv_file, 'r') as csvfile:
        reader = csv.reader(csvfile)
        next(reader)  # Skip the header row
        for row in reader:
            gpu_clocks.append(int(row[0]))
            max_powers.append(float(row[1]))

    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(gpu_clocks, max_powers, marker='o', linestyle='-', color='b')
    plt.title('GPU Clock Frequency vs Max Power Consumption')
    plt.xlabel('GPU Clock Frequency (MHz)')
    plt.ylabel('Max Power Consumption (W)')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    plot_results('gpu_power_results.csv')
```
