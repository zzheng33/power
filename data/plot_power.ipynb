```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_power_CPU(filename,benchmark):
    # Load the CSV file into a DataFrame
    data = pd.read_csv(filename)

    # Calculate node power as the sum of CPU and DRAM power
    data['Node Power (W)'] = data['Package Power (W)'] + data['DRAM Power (W)']

    # Plot the data
    plt.figure(figsize=(5, 3))
    plt.plot(data['Time (s)'], data['Package Power (W)'], label='RAPL Package', color='blue')
    plt.plot(data['Time (s)'], data['DRAM Power (W)'], label='RAPL DRAM', color='red')
    plt.plot(data['Time (s)'], data['Node Power (W)'], label='RAPL Node', color='green')  # Add node power line
    plt.xlabel('Time (s)')
    plt.ylabel('Power (W)')
    plt.title(f'Power Consumption ({benchmark})', pad=50)
    # Position the legend outside the plot
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.2), ncol=3)
    plt.grid(True,axis="y",alpha=0.5)
    plt.show()

```

```python
plot_power_CPU("./LULESH.csv","LULESH")
```

```python
plot_power_CPU("./nekbone.csv","Nekbone")
```

```python
plot_power_CPU("./miniFE.csv","miniFE")
```
