```python
import pandas as pd
import matplotlib.pyplot as plt
import os  # Import the os module to manage file path

def plot_power_CPU_GPU(cpu_filename, gpu_filename, benchmark):
    # Load the CSV files into DataFrames
    cpu_data = pd.read_csv(cpu_filename)
    gpu_data = pd.read_csv(gpu_filename)

    # Plot the data
    plt.figure(figsize=(6, 4))
    plt.plot(cpu_data['Time (s)'], cpu_data['Package Power (W)'], label='RAPL Package Power', color='blue')
    plt.plot(gpu_data['Time (s)'], gpu_data['GPU 0 Power (W)'], label='GPU Power', color='red')
    plt.xlabel('Time (s)')
    plt.ylabel('Power (W)')
    plt.title(f'Power Consumption ({benchmark})', pad=40)
    # Position the legend outside the plot
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=3)
    plt.grid(True, axis="y", alpha=0.5)
    
    # Make sure the ./fig directory exists, if not, create it
    os.makedirs('./fig', exist_ok=True)
    # Save the figure
    plt.savefig(f'./fig/power_cpu_gpu_{benchmark}.jpg', bbox_inches='tight')
    plt.show()



def plot_power_CPU(cpu_filename, benchmark):
    # Load the CSV files into DataFrames
    cpu_data = pd.read_csv(cpu_filename)

    # Plot the data
    plt.figure(figsize=(6, 4))
    plt.plot(cpu_data['Time (s)'], cpu_data['Package Power (W)'], label='RAPL Package Power', color='blue')
    plt.xlabel('Time (s)')
    plt.ylabel('Power (W)')
    plt.title(f'Power Consumption ({benchmark})', pad=40)
    # Position the legend outside the plot
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=3)
    plt.grid(True, axis="y", alpha=0.5)
    plt.show()
    # Make sure the ./fig directory exists, if not, create it
    os.makedirs('./fig', exist_ok=True)
    # Save the figure
    plt.savefig(f'./fig/power_cpu_{benchmark}.jpg', bbox_inches='tight')
    plt.close()  # Close the figure after saving to free up memory




def plot_power_cap(filename,benchmark):
    # Read data from CSV file
    data = pd.read_csv(filename)
    
    # Divide the data into three parts
    part1 = data[data['CPU Cap (W)'] == 125]  # First part: CPU cap is 125, GPU cap changes
    part2 = data[data['GPU Cap (W)'] == 260]  # Second part: GPU cap is 260, CPU cap changes

    # Plotting the data
    plt.figure(figsize=[6, 3])
    
    
    # Plot for the first part (CPU cap is consistent, GPU cap changes)
    plt.subplot(1, 2, 1)  # First subplot in a 1x2 grid
    plt.scatter(part1['GPU Cap (W)'], part1['Runtime (s)'])
    plt.title('GPU-cap only: '+benchmark)
    plt.xlabel('GPU Cap (W)')
    plt.ylabel('Runtime (s)')
    
    # Plot for the second part (GPU cap is consistent, CPU cap changes)
    plt.subplot(1, 2, 2)  # Second subplot in a 1x2 grid
    plt.scatter(part2['CPU Cap (W)'], part2['Runtime (s)'])
    plt.title('CPU-cap only: '+benchmark)
    plt.xlabel('CPU Cap (W)')
    plt.ylabel('Runtime (s)')
    
    plt.tight_layout()  # Adjust layout to not overlap
    plt.show()
```
